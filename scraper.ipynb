```python
##now combine the pages into one mega text
import codecs
pgs = glob("pages/page*.html")
pgs.sort()
open("SEwtHA.txt","w").write("")
for pg in pgs:
    text = open(pg).read()
    soup = BeautifulSoup(text)
    core = soup.find("div", {"id": "textcol"})
    text = core.get_text()
    text = text.replace("\r\n"," ")
    codecs.open("SEwtHA.txt",'a', "utf-8").write(text)
```

```python
##now combine the pages into one mega html
import codecs
pgs = glob("pages/page*.html")
pgs.sort()
open("SEwtHA.html","w").write("")
for pg in pgs:
    text = open(pg).read()
    soup = BeautifulSoup(text)
    core = soup.find("div", {"id": "textcol"})
    text = str(core.extract())
    text = text.replace("-<br/>\n","")
    text = text.replace("<br/>\n"," ")
    text = text.replace("-<br/>\r\n","")
    text = text.replace("<br/>\r\n"," ")
    text = text.decode("utf-8")
    codecs.open("SEwtHA.html",'a', "utf-8").write(text)
```
